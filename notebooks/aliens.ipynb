```python
import numpy as np
import h5py
import matplotlib.pyplot as plt
```

```python
with h5py.File('../data/block_o_spectra.hdf5') as f:
    fluxs = np.copy(f['fluxs'])
    waves = np.copy(f['waves'])
    star_names = np.copy(f['star_name'])
    
N = len(star_names)
```

One example of a potential Tc line:

```python
inds = (waves > 6387.) & (waves < 6392.)
for i in range(N):
    plt.plot(waves[inds], fluxs[i,inds])
plt.axvline(6389.88, color='k')
```
